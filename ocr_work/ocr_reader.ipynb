```python
# Step 2: Import libraries
import pytesseract
from PIL import Image
import os
```

```python
def extract():

    input_folder_path = './img/cropped-tif'  # Replace 'your_folder_name' with your folder's name
    output_folder_path = './cleaned-tif-cropped-TST'  # Replace with desired output folder name

    ################################
    # Step 5: Create the output folder if it doesn't exist

        
    if not os.path.exists(output_folder_path):
        os.makedirs(output_folder_path)
    
    ################################
    # Step 6: Loop through all images in the input folder
    for filename in os.listdir(input_folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif', 'tif')):  # Process only image files
            image_path = os.path.join(input_folder_path, filename)
            image = Image.open(image_path)

            image = image.convert('L')  # Grayscale conversion
            image = image.point(lambda x: 0 if x < 128 else 255, '1')  # Thresholding for binary format


            # Extract text using Tesseract
            custom_config = '--oem 1 -l econ+eng'  # Adjust based on layout

            extracted_text = pytesseract.image_to_string(image, config=custom_config)
            # Save the extracted text to a separate text file
            text_file_name = os.path.splitext(filename)[0] + '.txt'  # Use image name for text file name
            text_file_path = os.path.join(output_folder_path, text_file_name)

            with open(text_file_path, 'w', encoding='utf-8') as file:
                file.write(extracted_text)

            print(f"Text extracted from {filename} saved as {text_file_name}")

    print("Text extraction complete. All results are saved in the output folder.")
    return
```

```python
extract() 
print('\n\ndone with all extractions')
```

```python
from natsort import natsorted
import os
output_folder_path = './cleaned-tif-cropped-TST'

combined = './combined-txt-full-tif-cropped.txt'
total = ""
sorted = natsorted(os.listdir(output_folder_path))
for filename in sorted:
    with open (os.path.join(output_folder_path, filename), encoding='utf-8') as f:
        txt = f.read()
        total += txt
        total += "\n-------\nPAGE BREAK\n-------\n"

with open (combined, 'w') as f:
    f.write(total)
        
```
